```python
# Assuming 'df' is your DataFrame and 'word' is your tokenizer function
# Reset 'x' to the original password column
x = df["password"]

tdif = TfidfVectorizer(tokenizer=word)
x = tdif.fit_transform(x)  # Now apply fit_transform to the original text data
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
print(model.score(x_test,y_test))
```
